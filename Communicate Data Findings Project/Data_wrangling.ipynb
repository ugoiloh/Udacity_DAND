```python
# Convert the trip start and end times to timestamp datatype
trip.started_at = pd.to_datetime(trip.started_at, format = "%d/%m/%Y %H:%M")
trip.ended_at = pd.to_datetime(trip.ended_at, format = "%d/%m/%Y %H:%M")
```

```python
# Make some of the column headings more descriptive
trip.rename({'rideable_type': 'bike_type', 'member_casual': 'user_type',
             'started_at': 'start_time', 'ended_at': 'end_time'}, axis=1, inplace=True)
```

```python
# Making a concise table in which the visualizations would be based on
trip = trip[['bike_type', 'start_time', 'end_time', 'user_type']]
```

```python
# Calculating the time spent on each trip in minutes
trip_diff = trip['end_time'] - trip['start_time']
trip['trip_duration'] = trip_diff /np.timedelta64(1,'m')
```

```python
trip
```

```python
# Observing the trip duration statistics
trip.trip_duration.describe()
```

> Phew! A negative min value? We would explore further on this. The max value of 34350 suggests an outlier as the third quartile is just 27. 

```python
# Poking the data for insight into the negative min value
trip.query('trip_duration < 0')
```

> The exploration above suggests that the negative value is most probably an error due to swapping the start and end time. So, we would effect the correction.

```python
time_error = trip.query('trip_duration < 0')
for i, row in time_error.iterrows():
# Swap the columns
    trip.iloc[i,1] = row.end_time
    trip.iloc[i,2] = row.start_time

#Recalculating the trip duration
trip_diff = trip['end_time'] - trip['start_time']
trip['trip_duration'] = trip_diff /np.timedelta64(1,'m')
```

```python
# Confirm the swap was corrected done.
trip.query('trip_duration < 0')
```

```python
# Let's observe the suspected outlier
trip.query('trip_duration == 34350')
```

> Wow! 24 days on a trip of riding bike within a city? Sounds incredible. However, we would allow this as we have no information on the trip. 

```python
# Creating more table data from the trip start and end times for visualization

trip['start_time_weekday'] = trip['start_time'].dt.strftime('%A')

trip['start_time_hour'] = trip['start_time'].dt.hour
trip['end_time_hour'] = trip['end_time'].dt.hour
```

```python
# Let's observe the added table data
trip
```

```python
# sectioning the trip based on the period of the day
# {morning: 4-11, afternoon: 12-19, night: 20-23-3}
# The 'night1' label was to capture the other night range, which was corrected in the next cell
trip['start_day_period'] = pd.cut(x = trip['start_time_hour'], bins=[-1,3,11,19,23], 
                            labels=['night', 'morning', 'afternoon', 'night1'])
```

```python
# change the night1 variable to night
trip['start_day_period'] = trip['start_day_period'].apply(lambda x: x.replace('night1','night'))
```

```python
# checking to ensure it was correctly replaced
trip['start_day_period'].unique()
```

```python
# Change the user_type and bike_type to category dtype
trip.user_type = trip.user_type.astype('category')
trip.bike_type = trip.bike_type.astype('category')
```
